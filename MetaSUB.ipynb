```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r"F:\Annie\Cornell MS\MetaSUB\80per.csv")
df.head()
```

```python
df["total"] = df.sum(axis=1)
calc=0
for idx, row in df.iterrows():
    if  df.loc[idx,'total'] == 0:
        df.loc[idx,'total'] = 1
        print (idx)
        calc+=1
print(calc)
```

```python
df_new = df.loc[:,"Acetoanaerobium sticklandii":"uncultured crAssphage"].div(df["total"], axis=0)
df_new*100
```

```python

Total = df_new.sum()
non_zero = df_new.astype(bool).sum(axis=0)
plot = Total.plot.pie()
```

```python
Total.sort_values(axis=0, 
                 ascending=False, inplace=True)
non_zero.sort_values(axis=0, 
                 ascending=False, inplace=True)
length = (len(non_zero))
name_list = []
val_list = []
avg_count = []
i = 0
while (Total.iloc[i]>1):
    for j in range (length):
        if (non_zero.index[j]== Total.index[i]):
            if (non_zero[j]>calc):
                avg_count.append(non_zero[j]-calc)
                val_list.append(Total.iloc[i])
                name_list.append(Total.index[i])
    i = i+1
plt.bar(name_list,val_list)
```

```python
calc_avg = np.array(val_list, dtype=float)/np.array(avg_count)
plt.bar(name_list,calc_avg)
```

```python
length = (len(df_new.index)) - calc
actual_avg = np.array(val_list, dtype=float)/length
plt.bar(name_list,actual_avg)
```
